```python
"""
my own implementation of BO, with GPs and EI.
Varibles are in 1d, for visualisation purpose
"""
import numpy as np
import matplotlib.pyplot as plt
import copy

# store images in the notebook
%matplotlib inline 
%config InlineBackend.figure_format = 'svg' # use png for download image

from BO import bayes_optimiser

np.random.seed(0)
```

# environment

```python
class environment:
    def __init__(self):
        self.x = np.linspace(0,1,50) # use to smooth out the function
        self.x_init = np.array([0.3,0.55]) # initial samples
    
    def function(self,x):
        return np.sin(10*x) + np.exp(x)/2

    def transition(self,x):
        return self.function(x)
    
    def plot(self):
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 5))
        ax.plot(self.x,self.function(self.x))
        ax.set_title("the true function")
        ax.scatter(self.x_init,self.function(self.x_init),color='turquoise',label="initial samples")
        ax.legend()
        plt.show()

env = environment()
env.plot()
```

# bayesian optimisation

##### GP posterior


```python
# initial samples
x_sample = copy.deepcopy(env.x_init)
y_sample = env.function(x_sample)
X_test = np.linspace(0,1,100)

# hyper-parameter
hyper_params = {"kernel_type":"SE",
                "len_scale":5e-2,
                "amplify_coef":1e1,
                "xi":1e-2,
                }

bounds=np.array([[0,1]])
opt_restart = 10

bayes_opt = bayes_optimiser(**hyper_params)
bayes_opt.model.visualize(x_sample,y_sample,X_test,mean_function=0,mean_function_test=0,N_func_sample = 5)
```

##### acquisition function

```python
bayes_opt.plot_decision(bounds,opt_restart,x_sample,y_sample,X_test,mean_function=0,mean_function_test=0)
```

```python
_,x_next = bayes_opt.acquisition_function.propose_position(bayes_opt.model,bounds,opt_restart,x_sample,y_sample,X_test,mean_function=0,mean_function_test=0)
x_sample = np.append(x_sample,x_next)
y_sample = env.function(x_sample)
bayes_opt.plot_decision(bounds,opt_restart,x_sample,y_sample,X_test,mean_function=0,mean_function_test=0)
```
